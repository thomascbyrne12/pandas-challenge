```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources//purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_df = pd.DataFrame(purchase_data)
```

```python
purchase_df.head()
```

```python
total_players = pd.DataFrame({"Total Players": [purchase_df["Purchase ID"].count()]})
```

```python
total_players
```

```python
unique_items = purchase_df["Item Name"].nunique()
avg_price = purchase_df["Price"].mean()
total_purchases = purchase_df["Purchase ID"].count()
revenue = purchase_df["Price"].sum()
summary_df  = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price], 
                            "Number of Purchases": [total_purchases], "Total Revenue": [revenue]})
summary_df['Average Price'] = summary_df['Average Price'].map('${:,.2f}'.format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].map('${:,.2f}'.format)
summary_df.head()
```

```python
unique_df = purchase_df.dropna(how='any')
gender_total = unique_df["Gender"].value_counts()
percent = unique_df['Gender'].value_counts(normalize=True) * 100
gender_count = pd.DataFrame({"Total Count": purchase_df["Gender"].value_counts(), 
                             "Percentage of Players": percent})
gender_count['Percentage of Players'] = gender_count['Percentage of Players'].map('{:,.2f}%'.format)
gender_count
```

```python
gender_df = purchase_df.groupby("Gender")
gender_total = purchase_df["Gender"].value_counts()
gender_avg = gender_df["Price"].mean()
gender_price = gender_df["Price"].sum()
unique = purchase_df["Gender"].value_counts()
avg_per_person = gender_price / unique
gender_sort_df = pd.DataFrame({"Total Count": gender_total, "Average Purchase Price": gender_avg,
                                "Total Purchase Price": gender_price, 
                                "Average Purchase Price per Person": avg_per_person})
gender_sort_df['Average Purchase Price'] = gender_sort_df['Average Purchase Price'].map('${:,.2f}'.format)
gender_sort_df['Total Purchase Price'] = gender_sort_df['Total Purchase Price'].map('${:,.2f}'.format)
gender_sort_df['Average Purchase Price per Person'] = gender_sort_df['Average Purchase Price per Person'].map('${:,.2f}'.format)
gender_sort_df
```

```python
bins = [0,9,14,19,24,29,34,39,99]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
range_df = pd.cut(unique_df['Age'], bins = bins, labels = labels)
merge_df = pd.merge(purchase_df, range_df, left_index=True, right_index=True)
rename_df = merge_df.rename(columns = {'Age_x': 'Age', 'Age_y': 'Age Ranges'})
total = rename_df['Age Ranges'].value_counts()
percents = (rename_df['Age Ranges'].value_counts(normalize=True) * 100)
age_df = pd.DataFrame({'Total Count': total, 'Percentage of Players': percents})
age_df = age_df.reindex(['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])
age_df['Percentage of Players'] = age_df['Percentage of Players'].map('{:,.2f}%'.format)
age_df
```

```python
bins = [0,9,14,19,24,29,34,39,99]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
range_df = pd.cut(purchase_df['Age'], bins = bins, labels = labels)
merge_df = pd.merge(purchase_df, range_df, left_index=True, right_index=True)
rename_df = merge_df.rename(columns = {'Age_x': 'Age', 'Age_y': 'Age Ranges'})
group_df = rename_df.groupby('Age Ranges')
total_purchases = rename_df['Age Ranges'].value_counts()
revenue = group_df["Price"].sum()
unique = rename_df["Age Ranges"].value_counts()
avg_per_person = revenue / unique
avg_price = group_df["Price"].mean()
age_sort_df = pd.DataFrame({'Purchase Count': total_purchases, 'Average Purchase Price': avg_price, 
                            'Total Purchase Value': revenue, 'Average Total Purchase per Person': avg_per_person})
age_sort_df = age_sort_df.reindex(['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])
age_sort_df['Average Total Purchase per Person'] = age_sort_df['Average Total Purchase per Person'].map('${:,.2f}'.format)
age_sort_df['Total Purchase Value'] = age_sort_df['Total Purchase Value'].map('${:,.2f}'.format)
age_sort_df['Average Purchase Price'] = age_sort_df['Average Purchase Price'].map('${:,.2f}'.format)
age_sort_df
```

```python
name_df = purchase_df.groupby('SN')
avg_price = name_df["Price"].mean()
revenue = name_df["Price"].sum()
count = purchase_df['SN'].value_counts()
spenders_df = pd.DataFrame({'Purchase Count': count, 'Average Purchase Price': avg_price, 
                            'Total Purchase Value': revenue})
keys = ['Lisosia93', 'Idastidru52', 'Chamjask73', 'Iral74', 'Iskadarya95']
spenders_df = spenders_df.reindex(keys)
spenders_df['Total Purchase Value'] = spenders_df['Total Purchase Value'].map('${:,.2f}'.format)
spenders_df['Average Purchase Price'] = spenders_df['Average Purchase Price'].map('${:,.2f}'.format)
spenders_df.index = spenders_df.index.rename('SN')
spenders_df
```

```python
item_df = purchase_df.groupby(['Item ID', 'Item Name'])
purchases = purchase_df.groupby(['Item ID', 'Item Name']).size()
avg_purchase = item_df['Price'].mean()
total_purchase = item_df['Price'].sum()
item_group_df = pd.DataFrame({'Purchase Count': purchases,'Item Price': avg_purchase,
                              'Total Purchase Value': total_purchase})
item_group_df['Item Price'] = item_group_df['Item Price'].map('${:,.2f}'.format)
item_group_df['Total Purchase Value'] = item_group_df['Total Purchase Value'].map('${:,.2f}'.format)
item_group_df = item_group_df.sort_values('Purchase Count', ascending=False)
item_group_df.head()
```
